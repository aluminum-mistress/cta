```python
def get30minutedata(pair):
   
    df=pd.DataFrame(client.futures_historical_klines(pair,'30m',time_factor+'m UTC'))
                    
    df=df.iloc[:,:8]
    df.columns=['Time','Open','High','Low','Close','Volume','ATR','EMA']
    df=df.set_index("Time")
    df.index=pd.to_datetime(df.index, unit='ms')
    df=df.astype(float)
    
    df['ATR'] = ta.ATR(df['High'], df['Low'], df['Close'], timeperiod=ATR_length)
    df['EMA'] = ta.EMA(df['Close'], timeperiod=EMA_period)

    return df
```
